```python
from ultralytics import YOLO
model = YOLO("runs/detect/french_signs2/weights/last.pt") 

results = model.train(
    data="dataset_french_signs/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    freeze=10, 
    name="french_signs",
    mixup=0.1,
    device="mps",
    resume=True,
    cache=True
)
```

```python
results = model("streetview_images/6058367b-b429-4f34-8955-674ace44c18b.jpg")

# Display and save the results
for result in results:
    result.show()                  # Display result in a window
    result.save(filename="output.jpg") # Save the annotated image to disk
```
